```python
# Distribution of Happiness Score
plt.figure(figsize=(12, 6))
sns.histplot(data['Score'], bins=30)
plt.title('Distribution of Happiness Score')
plt.show()
```

```python
# Influence of GDP on Happiness Score
plt.figure(figsize=(12, 6))
sns.scatterplot(x='GDP per capita', y='Score', data=data)
plt.title('Influence of GDP per Capita on Happiness Score')
plt.show()
```

```python
# Influence of Social Support on Happiness Score
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Social support', y='Score', data=data)
plt.title('Influence of Social Support on Happiness Score')
plt.show()
```

```python
# Influence of Generosity on Happiness Score
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Generosity', y='Score', data=data)
plt.title('Influence of Social Support on Happiness Score')
plt.show()
```

```python
# Influence of Freedom to make life choices on Happiness Score
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Freedom to make life choices', y='Score', data=data)
plt.title('Influence of Social Support on Happiness Score')
plt.show()
```

```python
# Influence of Corruption on Happiness Score
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Perceptions of corruption', y='Score', data=data)
plt.title('Influence of Social Support on Happiness Score')
plt.show()
```

Based on the preliminary analysis, let's formulate a few hypotheses:

Hypothesis 1:
Hypothesis 2:
Hypothesis 3:

```python
# @title Default title text
# Test Hypothesis 1
corr_gdp_happiness = data['GDP per capita'].corr(data['Score'])
print(f'Correlation between GDP per Capita and Happiness Score: {corr_gdp_happiness}')
```

```python
# Test Hypothesis 2
corr_social_happiness = data['Social support'].corr(data['Score'])
print(f'Correlation between Social Support and Happiness Score: {corr_social_happiness}')

# Test Hypothesis 3
corr_corruption_happiness = data['Perceptions of corruption'].corr(data['Score'])
print(f'Correlation between Perceptions of Corruption and Happiness Score: {corr_corruption_happiness}')
```

```python
# @title Default title text
# Conclusions for Hypotheses
if corr_gdp_happiness > 0.5:
    print("Hypothesis 1 confirmed: Happiness Score is higher in countries with higher GDP per capita.")
else:
    print("Hypothesis 1 not confirmed.")

if corr_social_happiness > 0.5:
    print("Hypothesis 2 confirmed: Happiness Score is higher in countries with higher social support.")
else:
    print("Hypothesis 2 not confirmed.")

if corr_corruption_happiness < -0.5:
    print("Hypothesis 3 confirmed: Happiness Score is higher in countries with lower perceptions of corruption.")
else:
    print("Hypothesis 3 not confirmed.")
```
