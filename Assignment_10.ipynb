```python
#sales funnel
visited = (data['stage'] == 'Visited').sum()
added_to_cart = (data['stage'] == 'Added to Cart').sum()
purchased = (data['stage'] == 'Purchased').sum()
```

```python
print(visited)
print(added_to_cart)
print(purchased)
```

```python
#conversion
conversion_rate_visited_to_cart = (added_to_cart / visited) * 100
conversion_rate_cart_to_purchased = (purchased / added_to_cart) * 100
overall_conversion_rate = (purchased / visited) * 100

print(f"Visited to Added to Cart: {conversion_rate_visited_to_cart:.2f}%")
print(f"Added to Cart to Purchased: {conversion_rate_cart_to_purchased:.2f}%")
print(f"Visited to Purchased (Overall Conversion Rate): {overall_conversion_rate:.2f}%")
```
