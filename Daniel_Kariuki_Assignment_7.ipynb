```python
import pandas as pd
from pandas import *
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.stats as st
```

```python
from sklearn.datasets import load_iris

# import
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
iris_df.duplicated().sum()
```

```python
sb.pairplot(iris_df, hue="species")
plt.title('Pair Plot of Iris Species')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sb.boxplot(x='species', y='sepal length (cm)', data=iris_df)
plt.title('Sepal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
iris_df['species'] = pd.Categorical(iris_df['species']).codes
```

```python
# Correlation between features
# Calculate the correlation matrix
corr_matrix = iris_df.corr()

# Plotting
plt.figure(figsize=(10, 8))
sb.heatmap(corr_matrix, annot=True)
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```

```python
#Strong correlation is indicated by values closer to 1 or -1.
#For example, in the Iris dataset there is a strong positive correlation
#between petal length and petal width, as well as a strong negative correlation between sepal length and sepal width.
```
