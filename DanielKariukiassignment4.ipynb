```python
from bs4 import BeautifulSoup
from requests import get
import pandas as pd
from pandas import *
```

```python
link = get("https://www.goodreads.com/shelf/show/kenya")
```

```python
print(link.status_code)
print(link.text)
```

```python
soup = BeautifulSoup(link.text, "html.parser")
```

```python
if link.status_code ==200:
    soup = BeautifulSoup(link.text, "html.parser")
    # Find all <a> tags with class "bookTitle"
    books = soup.find_all("a", class_="bookTitle")
    #Find all <span> tags with itemprop "author"
    authors = soup.find_all("span", itemprop="author")

    #make list to store book titles and authors
    book_title = [book.get_text(strip=True) for book in books]

    author = [author.get_text(strip=True)for author in authors]

else:
    print(f"Error: status {link.status_code}")

#create dataframe
df = pd.DataFrame({'Book': book_title, 'Author': author})
print(df)
```
