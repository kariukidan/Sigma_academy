```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/modified_mineral_water_data.csv')
df.head()
```

```python
# View statistics on data
print(df.describe())
```

```python
print(df.dtypes)
```

```python
print(df.duplicated().sum())
```

```python
df = df.drop_duplicates()
print(df)
```

```python
# Visualization of calcium content depending on water pH
plt.figure(figsize=(10, 6))
sns.scatterplot(x='calcium_mg_per_L', y='pH', data=df)
plt.title('Calcium content and pH level')
plt.show()
```

```python
# Visualization of magnesium content depending on water pH
plt.figure(figsize=(10, 6))
sns.scatterplot(x='magnesium_mg_per_L', y='pH', data=df)
plt.title('Magnesium content and pH level')
plt.show()
```

```python
# Distribution of pH levels in mineral water
plt.figure(figsize=(10, 6))
sns.histplot(df['pH'], kde=True)
plt.title('PH level distribution')
plt.show()
```

```python
# Distribution of total dissolved solids (TDS)
plt.figure(figsize=(10, 6))
sns.histplot(df['TDS_mg_per_L'], kde=True)
plt.title('Distribution of total dissolved solids (TDS)')
plt.show()
```

```python
# Relationship between carbonation level and chloride content
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carbonation_level', y='chloride_mg_per_L', data=df)
plt.title('Carbon level and chloride content')
plt.show()
```
