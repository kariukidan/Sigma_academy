Check the normality of the distribution of height and weight of individual genders the check must be performed both graphically and mathematically.


Construct a heat-weight correlation chart for both sexes (four columns in total)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
data = pd.read_csv('/content/weightheight (3).csv')
```

```python
data.head()
data.describe()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
for gender in data['Gender'].unique():
  plt.figure(figsize=(8,4))
  plt.subplot(1, 2, 1)
  sns.histplot(data[data['Gender']== gender]['Height'], kde=True)
  plt.title(f'Plot for Height of {gender}')
```

```python
for gender in data['Gender'].unique():
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(data[data['Gender'] == gender]['Height'], kde=True)
    plt.title(f'Height Distribution for {gender}')
```

```python
plt.subplot(1, 2, 2)
stats.probplot(data[data['Gender'] == gender]['Height'], dist="norm", plot=plt)
plt.title(f'Q-Q Plot for Height of {gender}')
```

```python
for gender in data['Gender'].unique():
    plt.subplot(1, 2, 2)
    stats.probplot(data[data['Gender'] == gender]['Height'], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot for Height of {gender}')
```

```python
for gender in data['Gender'].unique():
    # Create a new figure for each gender
    plt.figure()
    stats.probplot(data[data['Gender'] == gender]['Height'], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot for Height of {gender}')

plt.show()  # Display the p
```
